```python
import pandas as pd
file=pd.ExcelFile("Ditry Data Sample.xlsx")

dirty_df=file.parse("Dirty 1",skiprows=2)
# Map segment names to their columns
segments = {
    "Consumer": ["Order ID", "Unnamed: 1", "Unnamed: 2", "Unnamed: 3", "Unnamed: 4"],
    "Corporate": ["Order ID", "Unnamed: 6", "Unnamed: 7", "Unnamed: 8", "Unnamed: 9"],
    "Home Office": ["Order ID", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13", "Unnamed: 14"]
}

ship_modes = ["First Class", "Same Day", "Second Class", "Standard Class"]

# Clean and reshape
cleaned = []

for segment,cols in segments.items():
    for i,ship_mode in enumerate(ship_modes):
        temp=pd.DataFrame({"Segment":segment,
                           "Ship Mode":ship_mode,
                           "Order ID":dirty_df[cols[0]],
                           "Sales":pd.to_numeric(dirty_df[cols[i+1]],errors="coerce")
                          })
        cleaned.append(temp)
        
final_df=pd.concat(cleaned,ignore_index=True)
final_df=final_df.dropna(subset="Sales")

final_df.to_excel("sales_data_analyzed.xlsx",index=False)
```
